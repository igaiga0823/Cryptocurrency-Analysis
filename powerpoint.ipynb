```python
data = response.json()

data["market_cap_chart"]["market_cap"]
```

```python
doing_data = []
index_data = []

for i, value in enumerate(data["market_cap_chart"]["market_cap"]):
    index_data.append(i)
    doing_data.append(value[1])

doing_data
```

```python
import matplotlib.pyplot as plt

# プロット
plt.figure(figsize=(10, 6))
plt.plot(index_data, doing_data, linestyle='-', color='b', label='Market Cap (USD)')

# グラフのタイトルとラベル
plt.title("Total Cryptocurrency Market Capitalization", fontsize=14)
plt.xlabel("Day", fontsize=12)
plt.ylabel("Market Capitalization (USD)", fontsize=12)

# 軸の目盛りを整える
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# グリッドと凡例
plt.grid(alpha=0.3)
plt.legend(fontsize=10)

# グラフの表示
plt.tight_layout()
plt.show()
```
