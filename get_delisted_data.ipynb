## coindeckoを用いて上場廃止になった銘柄のデータ取得

```python

delisted_data_symbol = {
    "ant": "aragon",
    "bond": "barnbridge",
    "cvp": "concentrated-voting-power",
    "drep": "drep-new",
    "vgx": "ethos",
    "for": "force-protocol",
    "kp3r": "keep3rv1",
    "loom": "loom-network-new",
    "mob": "mobster",
    "xmr": "monero",
    "multi": "multichain-fantom",
    "xem": "nem",
    "omg": "omisego",
    "ooki": "ooki",
    "waves": "waves",
    "pnt": "pnetwork",
    "pols": "polkastarter",
    "reef": "reef",
    "idrt": "rupiah-token",
    "unfi": "unifi-protocol-dao",
    "vai": "vibe-ai",
    "wnxm": "wrapped-nxm"
}

```

```python
delisted_data_symbol_value = [i for i in delisted_data_symbol.values()]
```

```python
delisted_data_symbol_value
```

```python
import requests
import json
from collections import defaultdict

save_value = defaultdict(defaultdict)

for i in range(len(delisted_data_symbol_value)):
    # CoinGecko APIのURL
    URL = "https://pro-api.coingecko.com/api/v3"
    id = delisted_data_symbol_value[i]
    API_KEY = "CG-dwNyTAJCUD9bsWowq39ybASb"
    api_url = "?x_cg_pro_api_key=" + API_KEY

    stock_data = defaultdict(list)

    # エンドポイントとパラメータ
    endpoint = "/coins/" + id + "/market_chart/range"
    url = URL + endpoint + api_url
    params = {
        "vs_currency": "usd",  # 通貨
        "from": "1704067200",
        "to": "1712707199",
    }

    price_value = []
    volume_value = []
    # APIリクエスト
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        # 日次価格データを抽出
        prices = data["prices"]  # 日次価格データ (UNIXタイムスタンプ, 価格)
        total_volumes = data["total_volumes"]
        for j in range(len(data["prices"])):
            price_value.append(prices[j][1])
            volume_value.append(total_volumes[j][1])
        stock_data["price"] = price_value
        stock_data["volume"] = volume_value
        save_value[delisted_data_symbol_value[i]] = stock_data
    else:
        print(f"Error: {response.status_code}, {response.text}")
```

```python
import json

with open("./all_delisted_2024_data.json", "w") as file:
    json.dump(save_value, file, indent=4)
```

```python
import requests

def get_symbols_by_base_currency(base_currency):
    url = "https://api.binance.com/api/v3/exchangeInfo"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        # 基軸通貨が指定のもの（例: USDT）のペアを取得
        symbols = [symbol['symbol'] for symbol in data['symbols'] if symbol['quoteAsset'] == base_currency]
        return symbols
    else:
        print(f"Error: {response.status_code}")
        return []

# USDTを基軸通貨とするペアを取得
usdt_pairs = get_symbols_by_base_currency("USDT")
print(usdt_pairs)
```
